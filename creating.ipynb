```python
import psycopg2
from psycopg2 import Error
from psycopg2.extras import NamedTupleCursor

def execute_query(query, fetch_result=False):
    try:
        connection = psycopg2.connect(
                        database="postgres", 
                        user='postgres',
                        password='12341234', 
                        host='localhost', 
                        port='5433'
                    )
        
        connection.autocommit = True
        cursor = connection.cursor(cursor_factory=NamedTupleCursor)
        cursor.execute(query)
        if fetch_result:
            return cursor.fetchall()
    except (Exception, Error) as error:
        print("Error while connecting to PostgreSQL", error)
    finally:
        if (connection):
            cursor.close()
            connection.close()
```

## Создаем таблицы и загружаем в них данные из файлов csv

```python
import psycopg2 as pg_driver

db = pg_driver.connect(
    database="music_service_project", 
    user='musician',
    password='music', 
    host='localhost', 
    port='5433'
)


def execute_queries(db, sql_commands):
    db.autocommit = True
    # тот же самый курсор
    with db.cursor() as cursor:
        # в цикле поочередно выполняем запросы
        for sql_command in sql_commands:
            print(sql_command)
            # тот же самый .execute() для выполнения запроса
            cursor.execute(sql_command)


sql_commands = ["DROP TABLE IF EXISTS users;",
                "DROP TABLE IF EXISTS tracks;", 
                "DROP TABLE IF EXISTS albums;", 
                "DROP TABLE IF EXISTS artists;", 
                "DROP TABLE IF EXISTS days;", 
                "DROP TABLE IF EXISTS status;",
                "DROP TABLE IF EXISTS genres;", 
                "DROP TABLE IF EXISTS types;",
                "DROP TABLE IF EXISTS releases_of_albums;",
                "DROP TABLE IF EXISTS payments;", 
                "DROP TABLE IF EXISTS subscriptions;"
                
                """CREATE TABLE users (
                         user_id    VARCHAR       NOT NULL,
                         name       VARCHAR       NOT NULL,
                         age        INT           NOT NULL,
                         city       VARCHAR       NOT NULL, 
                         date_of_registration DATE NOT NULL, 
                         status_id  INT           NOT NULL
                );
                """,
                """CREATE TABLE tracks (
                         user_id    VARCHAR       NOT NULL,
                         track      VARCHAR       NOT NULL,
                         artist_id  BIGINT        NOT NULL, 
                         genre_id   INT           NOT NULL, 
                         duration_sec INT         NOT NULL, 
                         day_id       INT         NOT NULL,
                         type_id      INT         NOT NULL,
                         album_id     VARCHAR     NOT NULL, 
                         popularity   INT         NOT NULL
                  );
                """, 
                """CREATE TABLE albums  (
                         album_id    VARCHAR       NOT NULL,
                         name        VARCHAR       NOT NULL,
                         billboard   VARCHAR       NOT NULL,
                         artist_id   BIGINT       NOT NULL, 
                         popularity  INT           NOT NULL, 
                         total_tracks INT          NOT NULL,
                         album_type   VARCHAR      NOT NULL, 
                         image_url    VARCHAR      NOT NULL
                );
                """,
                """CREATE TABLE artists (
                         artist_id    BIGINT      NOT NULL,
                         name         VARCHAR     NOT NULL,
                         age          INT         NOT NULL, 
                         number_of_albums   INT   NOT NULL, 
                         number_of_EPs      INT   NOT NULL, 
                         number_of_singles  INT   NOT NULL
                  );
                """, 
                 """CREATE TABLE days (
                         day_id      INT        NOT NULL, 
                         day_name    VARCHAR    NOT NULL
                  );
                """, 
                """CREATE TABLE status (
                         status_id      INT        NOT NULL, 
                         status_name    VARCHAR    NOT NULL
                  );
                """, 
                """CREATE TABLE genres (
                         genre_id      INT        NOT NULL, 
                         genre_name    VARCHAR    NOT NULL, 
                         origin_country VARCHAR   NOT NULL
                  );
                """,  
                """CREATE TABLE types (
                         type_id      INT        NOT NULL, 
                         type_name    VARCHAR    NOT NULL
                  );
                """,
                """CREATE TABLE releases_of_albums (
                         album_id     VARCHAR        NOT NULL, 
                         release_date TIMESTAMP           NOT NULL, 
                         releases_date_precision VARCHAR NOT NULL 
                  );
                """,
                """CREATE TABLE payments (
                         user_id           VARCHAR        NOT NULL, 
                         subscription_id   INT            NOT NULL, 
                         usage_period_months INT          NOT NULL, 
                         money_paid          INT          NOT NULL
                  );
                """,
                  """CREATE TABLE subscriptions (
                         subscription_id       INT        NOT NULL, 
                         type_of_subscription  VARCHAR    NOT NULL, 
                         monthly_fee           INT        NOT NULL
                  );
                """
                ]


execute_queries(db, sql_commands)
```
